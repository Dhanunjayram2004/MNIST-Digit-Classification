```python
import tensorflow as tf
from tensorflow.keras import layers, Sequential
import numpy as np


X = np.array([100, 200, 300, 400, 500], dtype=float)
y = np.array([3005, 6005, 9005, 12005, 15005], dtype=float)


model = Sequential([
    layers.Dense(units=1, input_shape=[1])
])


model.compile(optimizer=tf.keras.optimizers.Adam(0.1), loss='mean_squared_error')


print("Training starts...")
model.fit(X, y, epochs=1000, verbose=0)
print("Training finished!")


prediction = model.predict(np.array([[1000.0]]))
print(f"Prediction for 1000 likes: {prediction[0][0]} views")
```

```python
import torch
import tensorflow as tf
```

```python
from tensorflow.keras import layers,models
```

```python
minst=tf.keras.datasets.mnist
```

```python
(x_train,y_train),(x_test,y_test)=minst.load_data()
```

```python
x_train,x_test=x_train/255.0,x_test/255.0
```

```python
model=models.Sequential([layers.Flatten(input_shape=(28,28)),layers.Dense(128,activation='relu'),layers.Dense(10,activation='softmax')])
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=5)
```

```python
test_loss,test_acc=model.evaluate(x_test,y_test)
print(test_acc*100)
```
